```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de la curva de potencia para un generador eólico de 2000 W (velocidad del viento en m/s y potencia en W)
velocidad_viento = np.array([0, 3, 5, 7, 9, 11, 13, 15, 17, 20, 25])
potencia = np.array([0, 0, 200, 600, 1000, 1400, 1800, 2000, 2000, 2000, 0])  # Potencia en W

# Función para interpolar la curva de potencia
def potencia_generada(velocidad):
    return np.interp(velocidad, velocidad_viento, potencia)

# Simulación de la velocidad del viento por hora durante 1000 días para 5 prosumidores
prosumer_count = 5
dias = 1000
horas_por_dia = 24
total_horas = dias * horas_por_dia

# Modelo de velocidad del viento (m/s) con variaciones diarias
np.random.seed(42)  # Para reproducibilidad
velocidades_viento = np.random.normal(5, 2, (prosumer_count, total_horas))  # Media de 5 m/s con desviación estándar de 2 m/s
velocidades_viento = np.clip(velocidades_viento, 0, 25)  # Limitar las velocidades entre 0 y 25 m/s

# Simular variaciones diarias (más viento durante el día, menos durante la noche)
for p in range(prosumer_count):
    for i in range(dias):
        for h in range(horas_por_dia):
            if 6 <= h <= 18:  # Más viento durante el día
                velocidades_viento[p, i * horas_por_dia + h] += np.random.normal(1, 0.5)
            else:  # Menos viento durante la noche
                velocidades_viento[p, i * horas_por_dia + h] -= np.random.normal(1, 0.5)

velocidades_viento = np.clip(velocidades_viento, 0, 25)  # Limitar las velocidades entre 0 y 25 m/s

# Calcular la generación de potencia por hora (W) para cada prosumidor
generacion_potencia = np.array([potencia_generada(velocidades_viento[p]) for p in range(prosumer_count)])

# Reshape para obtener matriz de (5, 1000, 24)
generacion_potencia = generacion_potencia.reshape((prosumer_count, dias, horas_por_dia))

# Graficar 5 días de datos en gráficos separados para el primer prosumidor
dias_para_graficar = 5
plt.figure(figsize=(15, 10))
for i in range(dias_para_graficar):
    plt.subplot(dias_para_graficar, 1, i + 1)
    plt.plot(range(horas_por_dia), generacion_potencia[0, i], label=f'Día {i+1}')
    plt.title(f'Generación de Potencia por Hora - Día {i+1} - Prosumidor 1')
    plt.xlabel('Hora del Día')
    plt.ylabel('Potencia (W)')
    plt.grid(True)
    plt.tight_layout()

plt.show()

# Mostrar la generación de potencia del segundo día del primer prosumidor como ejemplo
print(generacion_potencia[0, 1])

savemat('/content/drive/MyDrive/Varios/UDENAR/MTE/Codigos/Data/solar_generation_4.mat', {'array':generacion_potencia})
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import savemat

# Datos de la curva de potencia para un generador eólico de 2000 W (velocidad del viento en m/s y potencia en W)
velocidad_viento = np.array([0, 3, 5, 7, 9, 11, 13, 15, 17, 20, 25])
potencia = np.array([0, 0, 200, 600, 1000, 1400, 1800, 2000, 2000, 2000, 0])  # Potencia en W

# Función para interpolar la curva de potencia
def potencia_generada(velocidad):
    return np.interp(velocidad, velocidad_viento, potencia)

# Simulación de la velocidad del viento por hora durante 1000 días para 5 prosumidores
prosumer_count = 5
dias = 5000
horas_por_dia = 24
total_horas = dias * horas_por_dia

# Modelo de velocidad del viento (m/s) con variaciones diarias
np.random.seed(42)  # Para reproducibilidad
velocidades_viento = np.random.normal(5, 2, (prosumer_count, total_horas))  # Media de 5 m/s con desviación estándar de 2 m/s
velocidades_viento = np.clip(velocidades_viento, 0, 25)  # Limitar las velocidades entre 0 y 25 m/s

# Simular variaciones diarias (más viento durante el día, menos durante la noche)
for p in range(prosumer_count):
    for i in range(dias):
        for h in range(horas_por_dia):
            if 6 <= h <= 18:  # Más viento durante el día
                velocidades_viento[p, i * horas_por_dia + h] += np.random.normal(1, 0.5)
            else:  # Menos viento durante la noche
                velocidades_viento[p, i * horas_por_dia + h] -= np.random.normal(1, 0.5)

velocidades_viento = np.clip(velocidades_viento, 0, 25)  # Limitar las velocidades entre 0 y 25 m/s

# Calcular la generación de potencia por hora (W) para cada prosumidor
generacion_potencia = np.array([potencia_generada(velocidades_viento[p]) for p in range(prosumer_count)])

# Reshape para obtener matriz de (5, 1000, 24)
generacion_potencia = generacion_potencia.reshape((prosumer_count, dias, horas_por_dia))

# Calcular el factor de escala para ajustar los picos de generación
pico_max_deseado = 500  # Pico máximo de demanda en W
pico_max_actual = 2000  # Pico máximo de generación en W
factor_escala = pico_max_deseado / pico_max_actual

# Ajustar la generación de potencia
generacion_potencia_ajustada = generacion_potencia * factor_escala

# Graficar 5 días de datos en gráficos separados para el primer prosumidor
dias_para_graficar = 5
plt.figure(figsize=(15, 10))
for i in range(dias_para_graficar):
    plt.subplot(dias_para_graficar, 1, i + 1)
    plt.plot(range(horas_por_dia), generacion_potencia_ajustada[0, i], label=f'Día {i+1}')
    plt.title(f'Generación de Potencia Ajustada por Hora - Día {i+1} - Prosumidor 1')
    plt.xlabel('Hora del Día')
    plt.ylabel('Potencia (W)')
    plt.grid(True)
    plt.tight_layout()

plt.show()

# Mostrar la generación de potencia ajustada del segundo día del primer prosumidor como ejemplo
print(generacion_potencia_ajustada[0, 1])
print(generacion_potencia_ajustada.shape)

savemat('/content/drive/MyDrive/Varios/UDENAR/MTE/Códigos Final/Data final/generation_final_1.mat', {'array':generacion_potencia_ajustada})
```
